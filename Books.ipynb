```python
book['Author'].unique().sum()

#George R. R. Martin accured 2 times 'George R. R. Martin' and 'George R.R. Martin' the diffrance is space
# and also 'J.K. Rowling' and 'J. K. Rowling'
```

```python
#changing to the correct format
book['Author'] = book['Author'].replace(['George R. R. Martin'],'George R.R. Martin')
book['Author'] = book['Author'].replace(['J. K. Rowling'],'J.K. Rowling')
```

```python
fig, ax= plt.subplots(figsize=(15,10))
sns.heatmap(book.corr(), annot=True)
```

```python
#Useers book rating 
#Seaborn

fig, ax= plt.subplots(figsize=(15,10))
sns.countplot(x=book['User Rating'],palette='Set2')
plt.title('User Book Rating Plot',fontsize=16)
plt.show()
```

```python
#Seaborn
#Year 2014 has the most review
#Reviews over the years 
fig, ax= plt.subplots(figsize=(15,10))
sns.lineplot(data=book, x="Year", y="Reviews")
plt.title('Reviews over the years line plot',fontsize=16)
```

```python
#Matplotlib
#Genre type
fig, ax= plt.subplots(figsize=(15,10))
plt.pie(book['Genre'].value_counts(),
        explode = [0,0.1],
        labels=['Fiction','Non Fiction']
        ,shadow=True)
plt.title('Genre Type Pie Chart',fontsize=16);
```

```python
fig, ax= plt.subplots(figsize=(15,10))
sns.scatterplot(data=book,
                x="User Rating", 
                y="Price")
plt.title('User Rating vs. Price',fontsize=16);
```

```python
fig, ax= plt.subplots(figsize=(15,10))
sns.lineplot(x="Year", y="User Rating",
             hue="Genre",
             data=book)
plt.title('The relationship between the year and user rating based on the genre',fontsize=16);
```
